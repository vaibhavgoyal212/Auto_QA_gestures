# Duplicate labels for the synthesised dataset

```python
import pandas as pd
import numpy as np
import re
```

# Appropriateness Labels:

```python
df = pd.read_csv('../Dataset/Appropriateness_Labels/mean_opinion_scores.csv')

Conds = ['Cond_BA', 'Cond_BT,' 'Cond_N', 'Cond_M', 'Cond_SA', 'Cond_SB', 'Cond_SC', 'Cond_SD', 'Cond_SE']
original_studyseqs = 40
augmented_studyseqs = 800
new_df_indices = []
new_df_data = []
for k in range(0, len(Conds)+1):
    for i in range(0, int(augmented_studyseqs/original_studyseqs)):
        for j in range(1, original_studyseqs+1):
            original_idx = j
            new_idx = (i*40)+j
            original_studyseq_index = df['Unnamed: 0'][(j-1) + (k*40)]
            new_studyseq_index = original_studyseq_index[:re.search(r'\d', original_studyseq_index).start()] + str(new_idx)
            new_df_indices = new_df_indices + [new_studyseq_index]
            print(original_studyseq_index, new_studyseq_index)
            new_df_data = new_df_data + [df['Mean_Appropriateness_Score'][(j-1) + (k*40)]]
new_df = pd.DataFrame(data=new_df_data, index=new_df_indices, columns=['Mean_Appropriateness_Score'])
new_df.to_csv('Appropriateness_Labels/mean_opinion_scores.csv')
    
```
